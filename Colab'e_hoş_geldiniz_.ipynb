```
def map_group(x):
    x = str(x).strip()

    if x == "HEALTHY":
        return "Healthy"
    elif x.startswith("CVI"):
        return "CVI"
    elif x in [
        "Basal cell carcinoma",
        "Squamous cell carcinoma",
        "Morbus Bowen",
        "Actinic keratosis",
        "Seborrheic keratosis"
    ]:
        return "Keratinocytic tumors"
    elif x == "Systemic sclerosis":
        return "Systemic sclerosis"
    else:
        return "Other"

df_all["disease_group"] = df_all["TYPE OF DISEASE"].apply(map_group)

print(df_all["disease_group"].value_counts())
```

```
import os
import numpy as np
import pandas as pd
from PIL import Image

from skimage.color import rgb2gray
from skimage.feature import graycomatrix, graycoprops
from skimage.filters import sobel, threshold_otsu
from skimage.measure import shannon_entropy

# -----------------------------
# 1) Tek görüntüden feature çıkaran fonksiyon
# -----------------------------
def extract_features(image_path):
    img = np.array(Image.open(image_path))

    # RGB ise grayscale'a çevir
    if img.ndim == 3:
        # Eğer kanallar aynıysa ilk kanalı al
        if np.all(img[:,:,0] == img[:,:,1]) and np.all(img[:,:,1] == img[:,:,2]):
            gray = img[:,:,0]
        else:
            gray = (rgb2gray(img) * 255).astype(np.uint8)
    else:
        gray = img.astype(np.uint8)

    # temel intensity özellikleri
    mean_intensity = np.mean(gray)
    std_intensity = np.std(gray)
    min_intensity = np.min(gray)
    max_intensity = np.max(gray)
    median_intensity = np.median(gray)

    # entropi
    entropy_val = shannon_entropy(gray)

    # kenar yoğunluğu
    edge_img = sobel(gray)
    edge_mean = np.mean(edge_img)
    edge_std = np.std(edge_img)

    # vessel-like area fraction (çok kaba başlangıç ölçütü)
    # parlak damar benzeri yapılar için Otsu threshold
    try:
        thr = threshold_otsu(gray)
        binary = gray > thr
        bright_area_fraction = binary.mean()
    except:
        bright_area_fraction = np.nan

    # GLCM için 8-bit grayscale
    gray_8 = gray.astype(np.uint8)

    glcm = graycomatrix(
        gray_8,
        distances=[1],
        angles=[0],
        levels=256,
        symmetric=True,
        normed=True
    )

    glcm_contrast = graycoprops(glcm, 'contrast')[0, 0]
    glcm_dissimilarity = graycoprops(glcm, 'dissimilarity')[0, 0]
    glcm_homogeneity = graycoprops(glcm, 'homogeneity')[0, 0]
    glcm_energy = graycoprops(glcm, 'energy')[0, 0]
    glcm_correlation = graycoprops(glcm, 'correlation')[0, 0]
    glcm_asm = graycoprops(glcm, 'ASM')[0, 0]

    return {
        "mean_intensity": mean_intensity,
        "std_intensity": std_intensity,
        "min_intensity": min_intensity,
        "max_intensity": max_intensity,
        "median_intensity": median_intensity,
        "entropy": entropy_val,
        "edge_mean": edge_mean,
        "edge_std": edge_std,
        "bright_area_fraction": bright_area_fraction,
        "glcm_contrast": glcm_contrast,
        "glcm_dissimilarity": glcm_dissimilarity,
        "glcm_homogeneity": glcm_homogeneity,
        "glcm_energy": glcm_energy,
        "glcm_correlation": glcm_correlation,
        "glcm_asm": glcm_asm
    }

# -----------------------------
# 2) Tüm df_final üzerinde çalıştır
# -----------------------------
feature_rows = []

for idx, row in df_final.iterrows():
    feats = extract_features(row["file_path"])
    feats["file_name"] = row["file_name"]
    feats["case_id"] = row["case_id"]
    feats["TYPE OF DISEASE"] = row["TYPE OF DISEASE"]
    feats["disease_group"] = row["disease_group"]
    feature_rows.append(feats)

df_features = pd.DataFrame(feature_rows)

print(df_features.shape)
df_features.head()
```

```
import pandas as pd
import numpy as np
from scipy.stats import kruskal, mannwhitneyu

feature_cols = [
    "mean_intensity",
    "std_intensity",
    "median_intensity",
    "entropy",
    "edge_mean",
    "edge_std",
    "bright_area_fraction",
    "glcm_contrast",
    "glcm_dissimilarity",
    "glcm_homogeneity",
    "glcm_energy",
    "glcm_correlation",
    "glcm_asm"
]

groups = ["Healthy", "CVI", "Keratinocytic tumors"]

results = []

for feat in feature_cols:
    data_groups = [
        df_features[df_features["disease_group"] == g][feat].dropna().values
        for g in groups
    ]

    # genel test
    kw_stat, kw_p = kruskal(*data_groups)

    # ikili testler
    p_h_vs_cvi = mannwhitneyu(
        df_features[df_features["disease_group"] == "Healthy"][feat].dropna(),
        df_features[df_features["disease_group"] == "CVI"][feat].dropna(),
        alternative="two-sided"
    ).pvalue

    p_h_vs_tumor = mannwhitneyu(
        df_features[df_features["disease_group"] == "Healthy"][feat].dropna(),
        df_features[df_features["disease_group"] == "Keratinocytic tumors"][feat].dropna(),
        alternative="two-sided"
    ).pvalue

    p_cvi_vs_tumor = mannwhitneyu(
        df_features[df_features["disease_group"] == "CVI"][feat].dropna(),
        df_features[df_features["disease_group"] == "Keratinocytic tumors"][feat].dropna(),
        alternative="two-sided"
    ).pvalue

    results.append({
        "feature": feat,
        "kw_p": kw_p,
        "p_Healthy_vs_CVI": p_h_vs_cvi,
        "p_Healthy_vs_Tumor": p_h_vs_tumor,
        "p_CVI_vs_Tumor": p_cvi_vs_tumor,
        "Healthy_mean": df_features[df_features["disease_group"] == "Healthy"][feat].mean(),
        "CVI_mean": df_features[df_features["disease_group"] == "CVI"][feat].mean(),
        "Tumor_mean": df_features[df_features["disease_group"] == "Keratinocytic tumors"][feat].mean()
    })

df_stats = pd.DataFrame(results).sort_values("kw_p")
df_stats
```

```
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import (
    accuracy_score,
    roc_auc_score,
    classification_report,
    confusion_matrix
)

# Kullanılacak feature'lar
feature_cols = [
    "mean_intensity",
    "std_intensity",
    "median_intensity",
    "entropy",
    "edge_mean",
    "edge_std",
    "bright_area_fraction",
    "glcm_contrast",
    "glcm_dissimilarity",
    "glcm_homogeneity",
    "glcm_energy",
    "glcm_correlation",
    "glcm_asm"
]

# Binary target oluştur
df_model = df_features.copy()
df_model["binary_target"] = df_model["disease_group"].apply(
    lambda x: 1 if x == "Keratinocytic tumors" else 0
)

print(df_model["binary_target"].value_counts())

X = df_model[feature_cols]
y = df_model["binary_target"]

# Stratified split
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.30,
    random_state=42,
    stratify=y
)

print("Train size:", X_train.shape, "Test size:", X_test.shape)

# -------------------------
# Logistic Regression
# -------------------------
logreg_pipe = Pipeline([
    ("imputer", SimpleImputer(strategy="median")),
    ("scaler", StandardScaler()),
    ("model", LogisticRegression(
        max_iter=5000,
        class_weight="balanced",
        random_state=42
    ))
])

logreg_pipe.fit(X_train, y_train)

y_pred_lr = logreg_pipe.predict(X_test)
y_prob_lr = logreg_pipe.predict_proba(X_test)[:, 1]

print("=== Logistic Regression ===")
print("Accuracy:", accuracy_score(y_test, y_pred_lr))
print("ROC-AUC:", roc_auc_score(y_test, y_prob_lr))
print("Confusion matrix:\n", confusion_matrix(y_test, y_pred_lr))
print(classification_report(y_test, y_pred_lr, digits=4))

# Coefficients
lr_model = logreg_pipe.named_steps["model"]
coef_df = pd.DataFrame({
    "feature": feature_cols,
    "coef": lr_model.coef_[0]
}).sort_values("coef", ascending=False)

print("\nLogistic Regression Coefficients:")
print(coef_df)

# -------------------------
# Random Forest
# -------------------------
rf_pipe = Pipeline([
    ("imputer", SimpleImputer(strategy="median")),
    ("model", RandomForestClassifier(
        n_estimators=300,
        max_depth=None,
        min_samples_split=4,
        min_samples_leaf=2,
        class_weight="balanced",
        random_state=42
    ))
])

rf_pipe.fit(X_train, y_train)

y_pred_rf = rf_pipe.predict(X_test)
y_prob_rf = rf_pipe.predict_proba(X_test)[:, 1]

print("\n=== Random Forest ===")
print("Accuracy:", accuracy_score(y_test, y_pred_rf))
print("ROC-AUC:", roc_auc_score(y_test, y_prob_rf))
print("Confusion matrix:\n", confusion_matrix(y_test, y_pred_rf))
print(classification_report(y_test, y_pred_rf, digits=4))

rf_model = rf_pipe.named_steps["model"]
rf_imp_df = pd.DataFrame({
    "feature": feature_cols,
    "importance": rf_model.feature_importances_
}).sort_values("importance", ascending=False)

print("\nRandom Forest Feature Importances:")
print(rf_imp_df)
```

```
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve

fpr_lr, tpr_lr, _ = roc_curve(y_test, y_prob_lr)
fpr_rf, tpr_rf, _ = roc_curve(y_test, y_prob_rf)

plt.figure(figsize=(7,5))
plt.plot(fpr_lr, tpr_lr, label=f"LogReg AUC = {roc_auc_score(y_test, y_prob_lr):.3f}")
plt.plot(fpr_rf, tpr_rf, label=f"RF AUC = {roc_auc_score(y_test, y_prob_rf):.3f}")
plt.plot([0,1], [0,1], linestyle="--")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("Tumor vs Non-tumor ROC")
plt.legend()
plt.tight_layout()
plt.show()
```

```
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import GroupShuffleSplit, StratifiedKFold, cross_val_score
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import (
    accuracy_score,
    roc_auc_score,
    classification_report,
    confusion_matrix,
    roc_curve
)

# =========================================================
# KLASÖR
# =========================================================
out_dir = "/content/derma_octa_project/results_patient_level"
os.makedirs(out_dir, exist_ok=True)

# =========================================================
# FEATURE SÜTUNLARI
# =========================================================
feature_cols = [
    "mean_intensity",
    "std_intensity",
    "median_intensity",
    "entropy",
    "edge_mean",
    "edge_std",
    "bright_area_fraction",
    "glcm_contrast",
    "glcm_dissimilarity",
    "glcm_homogeneity",
    "glcm_energy",
    "glcm_correlation",
    "glcm_asm"
]

# =========================================================
# PATIENT ID'YI df_features İLE BİRLEŞTİR
# df_all içinde PATIENT ID var; file_name üzerinden merge edeceğiz
# =========================================================
df_meta_for_merge = df_all[["file_name", "case_id", "PATIENT ID", "TYPE OF DISEASE", "disease_group"]].copy()
df_meta_for_merge["PATIENT ID"] = df_meta_for_merge["PATIENT ID"].astype(str).str.strip()

df_feat2 = df_features.merge(
    df_meta_for_merge[["file_name", "PATIENT ID"]],
    on="file_name",
    how="left"
)

df_feat2["PATIENT ID"] = df_feat2["PATIENT ID"].astype(str).str.strip()

print("df_feat2 shape:", df_feat2.shape)
print("Missing PATIENT ID:", df_feat2["PATIENT ID"].isna().sum())

# =========================================================
# BOOTSTRAP CI FONKSİYONU
# =========================================================
def bootstrap_auc_ci(y_true, y_prob, n_boot=2000, seed=42):
    rng = np.random.default_rng(seed)
    y_true = np.array(y_true)
    y_prob = np.array(y_prob)

    boot_aucs = []
    n = len(y_true)

    for _ in range(n_boot):
        idx = rng.integers(0, n, n)
        y_b = y_true[idx]
        p_b = y_prob[idx]

        # Her iki sınıf da olmalı
        if len(np.unique(y_b)) < 2:
            continue

        auc_b = roc_auc_score(y_b, p_b)
        boot_aucs.append(auc_b)

    lower = np.percentile(boot_aucs, 2.5)
    upper = np.percentile(boot_aucs, 97.5)
    return np.mean(boot_aucs), lower, upper, np.array(boot_aucs)

# =========================================================
# PATIENT-LEVEL SPLIT FONKSİYONU
# =========================================================
def patient_level_split(df, target_col, group_col="PATIENT ID", test_size=0.30, random_state=42):
    X = df[feature_cols].copy()
    y = df[target_col].copy()
    groups = df[group_col].copy()

    gss = GroupShuffleSplit(n_splits=1, test_size=test_size, random_state=random_state)
    train_idx, test_idx = next(gss.split(X, y, groups=groups))

    X_train = X.iloc[train_idx].copy()
    X_test = X.iloc[test_idx].copy()
    y_train = y.iloc[train_idx].copy()
    y_test = y.iloc[test_idx].copy()
    groups_train = groups.iloc[train_idx].copy()
    groups_test = groups.iloc[test_idx].copy()

    return X_train, X_test, y_train, y_test, groups_train, groups_test

# =========================================================
# MODELLER
# =========================================================
logreg_pipe = Pipeline([
    ("imputer", SimpleImputer(strategy="median")),
    ("scaler", StandardScaler()),
    ("model", LogisticRegression(
        max_iter=5000,
        class_weight="balanced",
        random_state=42
    ))
])

rf_pipe = Pipeline([
    ("imputer", SimpleImputer(strategy="median")),
    ("model", RandomForestClassifier(
        n_estimators=300,
        max_depth=None,
        min_samples_split=4,
        min_samples_leaf=2,
        class_weight="balanced",
        random_state=42
    ))
])

# =========================================================
# ANALİZ 1: TUMOR vs NON-TUMOR
# =========================================================
df_tn = df_feat2.copy()
df_tn["binary_target"] = df_tn["disease_group"].apply(lambda x: 1 if x == "Keratinocytic tumors" else 0)

print("\n===== Tumor vs Non-tumor =====")
print(df_tn["binary_target"].value_counts())

X_train, X_test, y_train, y_test, groups_train, groups_test = patient_level_split(
    df_tn, target_col="binary_target", group_col="PATIENT ID", test_size=0.30, random_state=42
)

print("Train shape:", X_train.shape, "Test shape:", X_test.shape)
print("Unique patients train:", groups_train.nunique(), "test:", groups_test.nunique())

# Logistic Regression
logreg_pipe.fit(X_train, y_train)
y_pred_lr = logreg_pipe.predict(X_test)
y_prob_lr = logreg_pipe.predict_proba(X_test)[:, 1]

lr_acc = accuracy_score(y_test, y_pred_lr)
lr_auc = roc_auc_score(y_test, y_prob_lr)
lr_cm = confusion_matrix(y_test, y_pred_lr)
lr_report = classification_report(y_test, y_pred_lr, digits=4)

lr_boot_mean, lr_ci_low, lr_ci_high, lr_boot_dist = bootstrap_auc_ci(y_test, y_prob_lr, n_boot=2000, seed=42)

lr_model = logreg_pipe.named_steps["model"]
coef_df = pd.DataFrame({
    "feature": feature_cols,
    "coef": lr_model.coef_[0],
    "abs_coef": np.abs(lr_model.coef_[0])
}).sort_values("abs_coef", ascending=False)

# Random Forest
rf_pipe.fit(X_train, y_train)
y_pred_rf = rf_pipe.predict(X_test)
y_prob_rf = rf_pipe.predict_proba(X_test)[:, 1]

rf_acc = accuracy_score(y_test, y_pred_rf)
rf_auc = roc_auc_score(y_test, y_prob_rf)
rf_cm = confusion_matrix(y_test, y_pred_rf)
rf_report = classification_report(y_test, y_pred_rf, digits=4)

rf_boot_mean, rf_ci_low, rf_ci_high, rf_boot_dist = bootstrap_auc_ci(y_test, y_prob_rf, n_boot=2000, seed=42)

rf_model = rf_pipe.named_steps["model"]
rf_imp_df = pd.DataFrame({
    "feature": feature_cols,
    "importance": rf_model.feature_importances_
}).sort_values("importance", ascending=False)

# ROC
fpr_lr, tpr_lr, _ = roc_curve(y_test, y_prob_lr)
fpr_rf, tpr_rf, _ = roc_curve(y_test, y_prob_rf)

plt.figure(figsize=(7,5))
plt.plot(fpr_lr, tpr_lr, label=f"LogReg AUC = {lr_auc:.3f}")
plt.plot(fpr_rf, tpr_rf, label=f"RF AUC = {rf_auc:.3f}")
plt.plot([0,1], [0,1], linestyle="--")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("Patient-level ROC: Tumor vs Non-tumor")
plt.legend()
plt.tight_layout()
plt.savefig(os.path.join(out_dir, "roc_patient_level_tumor_vs_nontumor.png"), dpi=300)
plt.show()

# =========================================================
# ANALİZ 2: HEALTHY vs CVI
# =========================================================
df_hc = df_feat2[df_feat2["disease_group"].isin(["Healthy", "CVI"])].copy()
df_hc["hc_target"] = df_hc["disease_group"].apply(lambda x: 1 if x == "CVI" else 0)

print("\n===== Healthy vs CVI =====")
print(df_hc["hc_target"].value_counts())

X_train_hc, X_test_hc, y_train_hc, y_test_hc, groups_train_hc, groups_test_hc = patient_level_split(
    df_hc, target_col="hc_target", group_col="PATIENT ID", test_size=0.30, random_state=42
)

print("Train shape:", X_train_hc.shape, "Test shape:", X_test_hc.shape)
print("Unique patients train:", groups_train_hc.nunique(), "test:", groups_test_hc.nunique())

logreg_pipe.fit(X_train_hc, y_train_hc)
y_pred_hc = logreg_pipe.predict(X_test_hc)
y_prob_hc = logreg_pipe.predict_proba(X_test_hc)[:, 1]

hc_acc = accuracy_score(y_test_hc, y_pred_hc)
hc_auc = roc_auc_score(y_test_hc, y_prob_hc)
hc_cm = confusion_matrix(y_test_hc, y_pred_hc)
hc_report = classification_report(y_test_hc, y_pred_hc, digits=4)

hc_boot_mean, hc_ci_low, hc_ci_high, hc_boot_dist = bootstrap_auc_ci(y_test_hc, y_prob_hc, n_boot=2000, seed=42)

# =========================================================
# PUBLICATION-READY TABLOLAR
# =========================================================
summary_df = pd.DataFrame([
    {
        "analysis": "Tumor vs Non-tumor",
        "model": "Logistic Regression",
        "accuracy": lr_acc,
        "roc_auc_test": lr_auc,
        "roc_auc_boot_mean": lr_boot_mean,
        "roc_auc_ci_lower": lr_ci_low,
        "roc_auc_ci_upper": lr_ci_high,
        "n_test": len(y_test),
        "n_positive_test": int(np.sum(y_test))
    },
    {
        "analysis": "Tumor vs Non-tumor",
        "model": "Random Forest",
        "accuracy": rf_acc,
        "roc_auc_test": rf_auc,
        "roc_auc_boot_mean": rf_boot_mean,
        "roc_auc_ci_lower": rf_ci_low,
        "roc_auc_ci_upper": rf_ci_high,
        "n_test": len(y_test),
        "n_positive_test": int(np.sum(y_test))
    },
    {
        "analysis": "Healthy vs CVI",
        "model": "Logistic Regression",
        "accuracy": hc_acc,
        "roc_auc_test": hc_auc,
        "roc_auc_boot_mean": hc_boot_mean,
        "roc_auc_ci_lower": hc_ci_low,
        "roc_auc_ci_upper": hc_ci_high,
        "n_test": len(y_test_hc),
        "n_positive_test": int(np.sum(y_test_hc))
    }
])

# Grup dağılım tablosu
group_table = df_feat2.groupby("disease_group").agg(
    n_images=("file_name", "count"),
    n_patients=("PATIENT ID", "nunique")
).reset_index()

# Feature özeti
feature_summary = df_feat2.groupby("disease_group")[feature_cols].agg(["mean", "std"])
feature_summary.to_csv(os.path.join(out_dir, "feature_summary_by_group.csv"))

# =========================================================
# DOSYALARA KAYDET
# =========================================================
summary_df.to_csv(os.path.join(out_dir, "summary_metrics_patient_level.csv"), index=False)
group_table.to_csv(os.path.join(out_dir, "group_distribution_patient_level.csv"), index=False)
coef_df.to_csv(os.path.join(out_dir, "logreg_coefficients_patient_level.csv"), index=False)
rf_imp_df.to_csv(os.path.join(out_dir, "rf_feature_importances_patient_level.csv"), index=False)

pd.DataFrame({"y_test": y_test, "y_prob_lr": y_prob_lr, "y_prob_rf": y_prob_rf}).to_csv(
    os.path.join(out_dir, "tumor_vs_nontumor_test_predictions.csv"), index=False
)

pd.DataFrame({"y_test_hc": y_test_hc, "y_prob_hc": y_prob_hc}).to_csv(
    os.path.join(out_dir, "healthy_vs_cvi_test_predictions.csv"), index=False
)

with open(os.path.join(out_dir, "patient_level_reports.txt"), "w", encoding="utf-8") as f:
    f.write("="*90 + "\n")
    f.write("GROUP DISTRIBUTION\n")
    f.write("="*90 + "\n")
    f.write(group_table.to_string(index=False))
    f.write("\n\n")

    f.write("="*90 + "\n")
    f.write("Tumor vs Non-tumor | Logistic Regression\n")
    f.write("="*90 + "\n")
    f.write(f"Accuracy: {lr_acc:.4f}\n")
    f.write(f"Test ROC-AUC: {lr_auc:.4f}\n")
    f.write(f"Bootstrap AUC mean: {lr_boot_mean:.4f}\n")
    f.write(f"95% CI: [{lr_ci_low:.4f}, {lr_ci_high:.4f}]\n")
    f.write(f"Confusion matrix:\n{lr_cm}\n\n")
    f.write("Classification report:\n")
    f.write(lr_report + "\n\n")
    f.write("Top coefficients:\n")
    f.write(coef_df.head(15).to_string(index=False))
    f.write("\n\n")

    f.write("="*90 + "\n")
    f.write("Tumor vs Non-tumor | Random Forest\n")
    f.write("="*90 + "\n")
    f.write(f"Accuracy: {rf_acc:.4f}\n")
    f.write(f"Test ROC-AUC: {rf_auc:.4f}\n")
    f.write(f"Bootstrap AUC mean: {rf_boot_mean:.4f}\n")
    f.write(f"95% CI: [{rf_ci_low:.4f}, {rf_ci_high:.4f}]\n")
    f.write(f"Confusion matrix:\n{rf_cm}\n\n")
    f.write("Classification report:\n")
    f.write(rf_report + "\n\n")
    f.write("Top importances:\n")
    f.write(rf_imp_df.head(15).to_string(index=False))
    f.write("\n\n")

    f.write("="*90 + "\n")
    f.write("Healthy vs CVI | Logistic Regression\n")
    f.write("="*90 + "\n")
    f.write(f"Accuracy: {hc_acc:.4f}\n")
    f.write(f"Test ROC-AUC: {hc_auc:.4f}\n")
    f.write(f"Bootstrap AUC mean: {hc_boot_mean:.4f}\n")
    f.write(f"95% CI: [{hc_ci_low:.4f}, {hc_ci_high:.4f}]\n")
    f.write(f"Confusion matrix:\n{hc_cm}\n\n")
    f.write("Classification report:\n")
    f.write(hc_report + "\n")

# =========================================================
# EKRANA KISA ÖZET
# =========================================================
print("\n" + "="*90)
print("PATIENT-LEVEL SUMMARY")
print("="*90)
print(summary_df)
print("\nTop Logistic Regression coefficients:")
print(coef_df.head(10))
print("\nTop Random Forest importances:")
print(rf_imp_df.head(10))
print("\nGroup distribution:")
print(group_table)

print("\nKaydedilen dosyalar:")
for fn in sorted(os.listdir(out_dir)):
    print("-", fn)
```

```
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from PIL import Image
from scipy import ndimage as ndi

from skimage.color import rgb2gray
from skimage.filters import threshold_otsu, frangi, sobel
from skimage.morphology import skeletonize, remove_small_objects, binary_closing, disk
from skimage.measure import label, regionprops, shannon_entropy
from skimage.feature import graycomatrix, graycoprops

# =========================================================
# Yardımcı fonksiyonlar
# =========================================================

def to_gray_uint8(img):
    if img.ndim == 3:
        if np.all(img[:, :, 0] == img[:, :, 1]) and np.all(img[:, :, 1] == img[:, :, 2]):
            gray = img[:, :, 0]
        else:
            gray = (rgb2gray(img) * 255).astype(np.uint8)
    else:
        gray = img.astype(np.uint8)
    return gray


def box_count_fractal_dimension(Z):
    Z = (Z > 0)
    if Z.sum() == 0:
        return np.nan

    def boxcount(img, k):
        S = np.add.reduceat(
            np.add.reduceat(img, np.arange(0, img.shape[0], k), axis=0),
            np.arange(0, img.shape[1], k), axis=1
        )
        return len(np.where((S > 0) & (S < k * k))[0])

    p = min(Z.shape)
    n = 2 ** int(np.floor(np.log2(p)))
    sizes = 2 ** np.arange(int(np.log2(n)), 1, -1)

    counts = []
    valid_sizes = []
    for size in sizes:
        c = boxcount(Z, size)
        if c > 0:
            counts.append(c)
            valid_sizes.append(size)

    if len(counts) < 2:
        return np.nan

    coeffs = np.polyfit(np.log(valid_sizes), np.log(counts), 1)
    return -coeffs[0]


def branch_and_end_points(skel):
    skel = skel.astype(np.uint8)
    kernel = np.array([[1,1,1],
                       [1,10,1],
                       [1,1,1]])
    conv = ndi.convolve(skel, kernel, mode='constant', cval=0)

    endpoints = np.sum(conv == 11)
    branchpoints = np.sum(conv >= 14)

    return endpoints, branchpoints


def estimate_mean_width(binary_vessels, skeleton):
    vessel_area = binary_vessels.sum()
    skel_len = skeleton.sum()
    if skel_len == 0:
        return np.nan
    return vessel_area / skel_len


def tortuosity_proxy(skeleton):
    lbl = label(skeleton)
    props = regionprops(lbl)

    vals = []
    for r in props:
        coords = r.coords
        if len(coords) < 10:
            continue

        y0, x0 = coords[0]
        y1, x1 = coords[-1]
        euclid = np.sqrt((y1 - y0)**2 + (x1 - x0)**2)
        path_len = len(coords)

        if euclid > 0:
            vals.append(path_len / euclid)

    if len(vals) == 0:
        return np.nan
    return float(np.mean(vals))


def vessel_segmentation(gray):
    fr = frangi(gray / 255.0)
    fr = (fr - fr.min()) / (fr.max() - fr.min() + 1e-8)

    try:
        thr = threshold_otsu(fr)
        binary = fr > thr
    except:
        binary = fr > 0.5

    binary = binary_closing(binary, footprint=disk(1))
    binary = remove_small_objects(binary, min_size=20)

    return fr, binary


def extract_vessel_features(image_path):
    img = np.array(Image.open(image_path))
    gray = to_gray_uint8(img)

    mean_intensity = np.mean(gray)
    std_intensity = np.std(gray)
    median_intensity = np.median(gray)
    entropy_val = shannon_entropy(gray)

    edge_img = sobel(gray)
    edge_mean = np.mean(edge_img)
    edge_std = np.std(edge_img)

    try:
        thr_gray = threshold_otsu(gray)
        bright_area_fraction = (gray > thr_gray).mean()
    except:
        bright_area_fraction = np.nan

    glcm = graycomatrix(
        gray.astype(np.uint8),
        distances=[1],
        angles=[0],
        levels=256,
        symmetric=True,
        normed=True
    )

    glcm_contrast = graycoprops(glcm, 'contrast')[0, 0]
    glcm_dissimilarity = graycoprops(glcm, 'dissimilarity')[0, 0]
    glcm_homogeneity = graycoprops(glcm, 'homogeneity')[0, 0]
    glcm_energy = graycoprops(glcm, 'energy')[0, 0]
    glcm_correlation = graycoprops(glcm, 'correlation')[0, 0]
    glcm_asm = graycoprops(glcm, 'ASM')[0, 0]

    frangi_img, vessel_bin = vessel_segmentation(gray)
    vessel_skel = skeletonize(vessel_bin)

    vessel_area_fraction = vessel_bin.mean()
    skeleton_length = vessel_skel.sum()
    skeleton_length_density = skeleton_length / vessel_skel.size

    endpoints, branchpoints = branch_and_end_points(vessel_skel)
    mean_width = estimate_mean_width(vessel_bin, vessel_skel)
    fractal_dim = box_count_fractal_dimension(vessel_bin)
    tortuosity = tortuosity_proxy(vessel_skel)

    lbl = label(vessel_bin)
    props = regionprops(lbl)
    n_components = len(props)

    if n_components > 0:
        component_areas = [r.area for r in props]
        largest_component_ratio = np.max(component_areas) / vessel_bin.sum() if vessel_bin.sum() > 0 else np.nan
    else:
        largest_component_ratio = np.nan

    return {
        "mean_intensity": mean_intensity,
        "std_intensity": std_intensity,
        "median_intensity": median_intensity,
        "entropy": entropy_val,
        "edge_mean": edge_mean,
        "edge_std": edge_std,
        "bright_area_fraction": bright_area_fraction,
        "glcm_contrast": glcm_contrast,
        "glcm_dissimilarity": glcm_dissimilarity,
        "glcm_homogeneity": glcm_homogeneity,
        "glcm_energy": glcm_energy,
        "glcm_correlation": glcm_correlation,
        "glcm_asm": glcm_asm,
        "vessel_area_fraction": vessel_area_fraction,
        "skeleton_length_density": skeleton_length_density,
        "branchpoints": branchpoints,
        "endpoints": endpoints,
        "mean_vessel_width": mean_width,
        "fractal_dimension": fractal_dim,
        "tortuosity_proxy": tortuosity,
        "n_components": n_components,
        "largest_component_ratio": largest_component_ratio
    }


feature_rows_v2 = []

for _, row in df_final.iterrows():
    feats = extract_vessel_features(row["file_path"])
    feats["file_name"] = row["file_name"]
    feats["case_id"] = row["case_id"]
    feats["TYPE OF DISEASE"] = row["TYPE OF DISEASE"]
    feats["disease_group"] = row["disease_group"]
    feature_rows_v2.append(feats)

df_features_v2 = pd.DataFrame(feature_rows_v2)

print("df_features_v2 shape:", df_features_v2.shape)
print(df_features_v2.head())
print("\nMissing values:")
print(df_features_v2.isna().sum().sort_values(ascending=False).head(15))

out_dir_v2 = "/content/derma_octa_project/results_vessel_features"
os.makedirs(out_dir_v2, exist_ok=True)

df_features_v2.to_csv(os.path.join(out_dir_v2, "derma_octa_features_v2.csv"), index=False)

print("\nKaydedildi:", os.path.join(out_dir_v2, "derma_octa_features_v2.csv"))
```

```
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import GroupShuffleSplit
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import roc_curve, roc_auc_score

base_dir = "/content/derma_octa_project/final_dataset_package"
fig_dir = f"{base_dir}/figures"
os.makedirs(fig_dir, exist_ok=True)

# Verileri oku
data_table_main = pd.read_csv(f"{base_dir}/data_table_main.csv")
features_v1 = pd.read_csv(f"{base_dir}/features_v1_final.csv")
features_v2 = pd.read_csv(f"{base_dir}/features_v2_final.csv")

# Target
features_v1["binary_target"] = features_v1["disease_group"].apply(lambda x: 1 if x == "Keratinocytic tumors" else 0)
features_v2["binary_target"] = features_v2["disease_group"].apply(lambda x: 1 if x == "Keratinocytic tumors" else 0)

feature_cols_v1 = [
    "mean_intensity", "std_intensity", "median_intensity", "entropy",
    "edge_mean", "edge_std", "bright_area_fraction",
    "glcm_contrast", "glcm_dissimilarity", "glcm_homogeneity",
    "glcm_energy", "glcm_correlation", "glcm_asm"
]

feature_cols_v2 = [
    "mean_intensity", "std_intensity", "median_intensity", "entropy",
    "edge_mean", "edge_std", "bright_area_fraction",
    "glcm_contrast", "glcm_dissimilarity", "glcm_homogeneity",
    "glcm_energy", "glcm_correlation", "glcm_asm",
    "vessel_area_fraction", "skeleton_length_density", "branchpoints",
    "endpoints", "mean_vessel_width", "fractal_dimension",
    "tortuosity_proxy", "n_components", "largest_component_ratio"
]

# Aynı patient-level split
groups = features_v1["PATIENT ID"].astype(str)
gss = GroupShuffleSplit(n_splits=1, test_size=0.30, random_state=42)
train_idx, test_idx = next(gss.split(features_v1[["file_name"]], features_v1["binary_target"], groups=groups))

def fit_and_roc(df, feature_cols, model_type="logreg"):
    X = df[feature_cols]
    y = df["binary_target"]

    X_train, X_test = X.iloc[train_idx], X.iloc[test_idx]
    y_train, y_test = y.iloc[train_idx], y.iloc[test_idx]

    if model_type == "logreg":
        pipe = Pipeline([
            ("imputer", SimpleImputer(strategy="median")),
            ("scaler", StandardScaler()),
            ("model", LogisticRegression(max_iter=5000, class_weight="balanced", random_state=42))
        ])
    else:
        pipe = Pipeline([
            ("imputer", SimpleImputer(strategy="median")),
            ("model", RandomForestClassifier(
                n_estimators=300,
                min_samples_split=4,
                min_samples_leaf=2,
                class_weight="balanced",
                random_state=42
            ))
        ])

    pipe.fit(X_train, y_train)
    y_prob = pipe.predict_proba(X_test)[:, 1]
    auc = roc_auc_score(y_test, y_prob)
    fpr, tpr, _ = roc_curve(y_test, y_prob)
    return fpr, tpr, auc

fpr1, tpr1, auc1 = fit_and_roc(features_v1, feature_cols_v1, "logreg")
fpr2, tpr2, auc2 = fit_and_roc(features_v1, feature_cols_v1, "rf")
fpr3, tpr3, auc3 = fit_and_roc(features_v2, feature_cols_v2, "logreg")
fpr4, tpr4, auc4 = fit_and_roc(features_v2, feature_cols_v2, "rf")

plt.figure(figsize=(7,6))
plt.plot(fpr1, tpr1, label=f"V1 Logistic (AUC={auc1:.3f})")
plt.plot(fpr2, tpr2, label=f"V1 Random Forest (AUC={auc2:.3f})")
plt.plot(fpr3, tpr3, label=f"V2 Logistic (AUC={auc3:.3f})")
plt.plot(fpr4, tpr4, label=f"V2 Random Forest (AUC={auc4:.3f})")
plt.plot([0, 1], [0, 1], linestyle="--")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("Figure 2. Patient-level ROC curves for tumor vs non-tumor classification")
plt.legend()
plt.tight_layout()
plt.savefig(f"{fig_dir}/Figure2_ROC_curves.png", dpi=300, bbox_inches="tight")
plt.show()
```

```
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from PIL import Image
from skimage.color import rgb2gray
from skimage.filters import threshold_otsu, frangi
from skimage.morphology import skeletonize, remove_small_objects, binary_closing, disk

base_dir = "/content/derma_octa_project/final_dataset_package"
fig_dir = f"{base_dir}/figures"
os.makedirs(fig_dir, exist_ok=True)

data_table_main = pd.read_csv(f"{base_dir}/data_table_main.csv")

def to_gray_uint8(img):
    if img.ndim == 3:
        if np.all(img[:, :, 0] == img[:, :, 1]) and np.all(img[:, :, 1] == img[:, :, 2]):
            gray = img[:, :, 0]
        else:
            gray = (rgb2gray(img) * 255).astype(np.uint8)
    else:
        gray = img.astype(np.uint8)
    return gray

def vessel_segmentation(gray):
    fr = frangi(gray / 255.0)
    fr = (fr - fr.min()) / (fr.max() - fr.min() + 1e-8)
    try:
        thr = threshold_otsu(fr)
        binary = fr > thr
    except:
        binary = fr > 0.5

    binary = binary_closing(binary, footprint=disk(1))
    binary = remove_small_objects(binary, min_size=20)
    skel = skeletonize(binary)
    return binary, skel

samples = []
for grp in ["Healthy", "CVI", "Keratinocytic tumors"]:
    temp = data_table_main[data_table_main["disease_group"] == grp].sample(2, random_state=42)
    samples.append(temp)

samples_df = pd.concat(samples, axis=0).reset_index(drop=True)

fig, axes = plt.subplots(len(samples_df), 3, figsize=(10, 3 * len(samples_df)))

for i, row in samples_df.iterrows():
    img = np.array(Image.open(row["file_path"]))
    gray = to_gray_uint8(img)
    vessel_mask, skel = vessel_segmentation(gray)

    axes[i, 0].imshow(gray, cmap="gray")
    axes[i, 0].set_title(f"{row['disease_group']}\nOriginal")
    axes[i, 0].axis("off")

    axes[i, 1].imshow(vessel_mask, cmap="gray")
    axes[i, 1].set_title("Vessel mask")
    axes[i, 1].axis("off")

    axes[i, 2].imshow(skel, cmap="gray")
    axes[i, 2].set_title("Skeleton")
    axes[i, 2].axis("off")

plt.suptitle("Figure 4. Representative OCTA images, vessel masks, and skeletonized vascular maps", y=1.0)
plt.tight_layout()
plt.savefig(f"{fig_dir}/Figure4_examples_masks_skeletons.png", dpi=300, bbox_inches="tight")
plt.show()
```

```
import os
import numpy as np
import pandas as pd

from sklearn.model_selection import GroupShuffleSplit
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import (
    roc_auc_score,
    average_precision_score,
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    balanced_accuracy_score,
    confusion_matrix
)

# =========================================================
# PATHS
# =========================================================
base_dir = "/content/derma_octa_project/final_dataset_package"
out_dir = f"{base_dir}/repeated_validation"
os.makedirs(out_dir, exist_ok=True)

# =========================================================
# LOAD DATA
# =========================================================
data_table_main = pd.read_csv(f"{base_dir}/data_table_main.csv")
features_v1 = pd.read_csv(f"{base_dir}/features_v1_final.csv")
features_v2 = pd.read_csv(f"{base_dir}/features_v2_final.csv")

# The 'PATIENT ID' column is already present in features_v1 and features_v2 from the previous steps.
# No need to merge again, but ensure it's treated as string and stripped.
features_v1["PATIENT ID"] = features_v1["PATIENT ID"].astype(str).str.strip()
features_v2["PATIENT ID"] = features_v2["PATIENT ID"].astype(str).str.strip()

# binary target
features_v1["binary_target"] = features_v1["disease_group"].apply(
    lambda x: 1 if x == "Keratinocytic tumors" else 0
)
features_v2["binary_target"] = features_v2["disease_group"].apply(
    lambda x: 1 if x == "Keratinocytic tumors" else 0
)

# =========================================================
# FEATURE COLUMNS
# =========================================================
feature_cols_v1 = [
    "mean_intensity", "std_intensity", "median_intensity", "entropy",
    "edge_mean", "edge_std", "bright_area_fraction",
    "glcm_contrast", "glcm_dissimilarity", "glcm_homogeneity",
    "glcm_energy", "glcm_correlation", "glcm_asm"
]

feature_cols_v2 = [
    "mean_intensity", "std_intensity", "median_intensity", "entropy",
    "edge_mean", "edge_std", "bright_area_fraction",
    "glcm_contrast", "glcm_dissimilarity", "glcm_homogeneity",
    "glcm_energy", "glcm_correlation", "glcm_asm",
    "vessel_area_fraction", "skeleton_length_density", "branchpoints",
    "endpoints", "mean_vessel_width", "fractal_dimension",
    "tortuosity_proxy", "n_components", "largest_component_ratio"
]

# =========================================================
# HELPERS
# =========================================================
def specificity_score(y_true, y_pred):
    cm = confusion_matrix(y_true, y_pred, labels=[0, 1])
    tn, fp, fn, tp = cm.ravel()
    return tn / (tn + fp) if (tn + fp) > 0 else np.nan

def sensitivity_score(y_true, y_pred):
    cm = confusion_matrix(y_true, y_pred, labels=[0, 1])
    tn, fp, fn, tp = cm.ravel()
    return tp / (tp + fn) if (tp + fn) > 0 else np.nan

def evaluate_once(df, feature_cols, model_name, seed, threshold=0.5):
    X = df[feature_cols].copy()
    y = df["binary_target"].copy()
    groups = df["PATIENT ID"].copy()

    gss = GroupShuffleSplit(n_splits=1, test_size=0.30, random_state=seed)
    train_idx, test_idx = next(gss.split(X, y, groups=groups))

    X_train, X_test = X.iloc[train_idx], X.iloc[test_idx]
    y_train, y_test = y.iloc[train_idx], y.iloc[test_idx]

    # test set must include both classes
    if len(np.unique(y_test)) < 2:
        return None

    if model_name == "Logistic Regression":
        model = Pipeline([
            ("imputer", SimpleImputer(strategy="median")),
            ("scaler", StandardScaler()),
            ("model", LogisticRegression(
                max_iter=5000,
                class_weight="balanced",
                random_state=seed
            ))
        ])
    elif model_name == "Random Forest":
        model = Pipeline([
            ("imputer", SimpleImputer(strategy="median")),
            ("model", RandomForestClassifier(
                n_estimators=300,
                min_samples_split=4,
                min_samples_leaf=2,
                class_weight="balanced",
                random_state=seed
            ))
        ])
    else:
        raise ValueError("Unknown model_name")

    model.fit(X_train, y_train)
    y_prob = model.predict_proba(X_test)[:, 1]
    y_pred = (y_prob >= threshold).astype(int)

    res = {
        "seed": seed,
        "model": model_name,
        "threshold": threshold,
        "n_train_images": len(train_idx),
        "n_test_images": len(test_idx),
        "n_train_patients": groups.iloc[train_idx].nunique(),
        "n_test_patients": groups.iloc[test_idx].nunique(),
        "n_test_positive": int(np.sum(y_test)),
        "roc_auc": roc_auc_score(y_test, y_prob),
        "pr_auc": average_precision_score(y_test, y_prob),
        "accuracy": accuracy_score(y_test, y_pred),
        "balanced_accuracy": balanced_accuracy_score(y_test, y_pred),
        "precision": precision_score(y_test, y_pred, zero_division=0),
        "recall": recall_score(y_test, y_pred, zero_division=0),
        "sensitivity": sensitivity_score(y_test, y_pred),
        "specificity": specificity_score(y_test, y_pred),
        "f1": f1_score(y_test, y_pred, zero_division=0),
    }
    return res

def run_repeated_validation(df, feature_cols, feature_set_name, n_runs=50, threshold=0.5):
    rows = []
    for seed in range(1, n_runs + 1):
        for model_name in ["Logistic Regression", "Random Forest"]:
            out = evaluate_once(
                df=df,
                feature_cols=feature_cols,
                model_name=model_name,
                seed=seed,
                threshold=threshold
            )
            if out is not None:
                out["feature_set"] = feature_set_name
                rows.append(out)
    return pd.DataFrame(rows)

def summarize_results(df_results):
    metrics = [
        "roc_auc", "pr_auc", "accuracy", "balanced_accuracy",
        "precision", "recall", "sensitivity", "specificity", "f1"
    ]
    rows = []
    for (feature_set, model), temp in df_results.groupby(["feature_set", "model"]):
        row = {
            "feature_set": feature_set,
            "model": model,
            "n_runs": len(temp),
            "median_test_positive": temp["n_test_positive"].median()
        }
        for m in metrics:
            row[f"{m}_mean"] = temp[m].mean()
            row[f"{m}_sd"] = temp[m].std()
            row[f"{m}_median"] = temp[m].median()
            row[f"{m}_q1"] = temp[m].quantile(0.25)
            row[f"{m}_q3"] = temp[m].quantile(0.75)
        rows.append(row)
    return pd.DataFrame(rows)

# =========================================================
# RUN 50-REPEAT VALIDATION
# =========================================================
results_v1 = run_repeated_validation(
    df=features_v1,
    feature_cols=feature_cols_v1,
    feature_set_name="V1_baseline",
    n_runs=50,
    threshold=0.5
)

results_v2 = run_repeated_validation(
    df=features_v2,
    feature_cols=feature_cols_v2,
    feature_set_name="V2_vessel_aware",
    n_runs=50,
    threshold=0.5
)

all_results = pd.concat([results_v1, results_v2], axis=0, ignore_index=True)
summary_results = summarize_results(all_results)

# =========================================================
# SAVE
# =========================================================
results_v1.to_csv(f"{out_dir}/repeated_results_v1.csv", index=False)
results_v2.to_csv(f"{out_dir}/repeated_results_v2.csv", index=False)
all_results.to_csv(f"{out_dir}/repeated_results_all.csv", index=False)
summary_results.to_csv(f"{out_dir}/repeated_results_summary.csv", index=False)

# =========================================================
# PRINT
# =========================================================
pd.set_option("display.max_columns", 200)

print("=== RAW RESULTS SHAPE ===")
print("V1:", results_v1.shape)
print("V2:", results_v2.shape)
print("ALL:", all_results.shape)
print()

print("=== SUMMARY RESULTS ===")
print(summary_results)

print()
print("Saved files:")
print(sorted(os.listdir(out_dir)))
```

```
import os
import numpy as np
import pandas as pd

from sklearn.model_selection import GroupShuffleSplit
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import (
    roc_auc_score,
    average_precision_score,
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    balanced_accuracy_score,
    confusion_matrix
)

base_dir = "/content/derma_octa_project/final_dataset_package"
out_dir = f"{base_dir}/threshold_analysis"
os.makedirs(out_dir, exist_ok=True)

features_v2 = pd.read_csv(f"{base_dir}/features_v2_final.csv")
data_table_main = pd.read_csv(f"{base_dir}/data_table_main.csv")

# The 'PATIENT ID' column is already present in features_v2_final.csv.
# No need to merge again, but ensure it's treated as string and stripped.
features_v2["PATIENT ID"] = features_v2["PATIENT ID"].astype(str).str.strip()
features_v2["binary_target"] = features_v2["disease_group"].apply(
    lambda x: 1 if x == "Keratinocytic tumors" else 0
)

feature_cols_v2 = [
    "mean_intensity", "std_intensity", "median_intensity", "entropy",
    "edge_mean", "edge_std", "bright_area_fraction",
    "glcm_contrast", "glcm_dissimilarity", "glcm_homogeneity",
    "glcm_energy", "glcm_correlation", "glcm_asm",
    "vessel_area_fraction", "skeleton_length_density", "branchpoints",
    "endpoints", "mean_vessel_width", "fractal_dimension",
    "tortuosity_proxy", "n_components", "largest_component_ratio"
]

def specificity_score(y_true, y_pred):
    cm = confusion_matrix(y_true, y_pred, labels=[0, 1])
    tn, fp, fn, tp = cm.ravel()
    return tn / (tn + fp) if (tn + fp) > 0 else np.nan

def run_thresholds(df, thresholds=[0.3, 0.5, 0.7], n_runs=30):
    rows = []
    X = df[feature_cols_v2]
    y = df["binary_target"]
    groups = df["PATIENT ID"]

    for seed in range(1, n_runs + 1):
        gss = GroupShuffleSplit(n_splits=1, test_size=0.30, random_state=seed)
        train_idx, test_idx = next(gss.split(X, y, groups=groups))

        X_train, X_test = X.iloc[train_idx], X.iloc[test_idx]
        y_train, y_test = y.iloc[train_idx], y.iloc[test_idx]

        if len(np.unique(y_test)) < 2:
            continue

        model = Pipeline([
            ("imputer", SimpleImputer(strategy="median")),
            ("scaler", StandardScaler()),
            ("model", LogisticRegression(
                max_iter=5000,
                class_weight="balanced",
                random_state=seed
            ))
        ])

        model.fit(X_train, y_train)
        y_prob = model.predict_proba(X_test)[:, 1]

        for thr in thresholds:
            y_pred = (y_prob >= thr).astype(int)
            rows.append({
                "seed": seed,
                "threshold": thr,
                "n_test_positive": int(np.sum(y_test)),
                "roc_auc": roc_auc_score(y_test, y_prob),
                "pr_auc": average_precision_score(y_test, y_prob),
                "accuracy": accuracy_score(y_test, y_pred),
                "balanced_accuracy": balanced_accuracy_score(y_test, y_pred),
                "precision": precision_score(y_test, y_pred, zero_division=0),
                "recall": recall_score(y_test, y_pred, zero_division=0),
                "specificity": specificity_score(y_test, y_pred),
                "f1": f1_score(y_test, y_pred, zero_division=0),
            })
    return pd.DataFrame(rows)

thr_results = run_thresholds(features_v2, thresholds=[0.3, 0.5, 0.7], n_runs=30)
thr_summary = thr_results.groupby("threshold").agg(
    roc_auc_mean=("roc_auc", "mean"),
    pr_auc_mean=("pr_auc", "mean"),
    accuracy_mean=("accuracy", "mean"),
    balanced_accuracy_mean=("balanced_accuracy", "mean"),
    precision_mean=("precision", "mean"),
    recall_mean=("recall", "mean"),
    specificity_mean=("specificity", "mean"),
    f1_mean=("f1", "mean")
).reset_index()

thr_results.to_csv(f"{out_dir}/threshold_results_v2_logreg.csv", index=False)
thr_summary.to_csv(f"{out_dir}/threshold_summary_v2_logreg.csv", index=False)

print(thr_summary)
```
